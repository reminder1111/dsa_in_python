Hint =>
a. The LeapYear program only works for year >= 1582, corresponding to a year in the Gregorian calendar. So ensure to check for the same.
b. Further, the Leap Year is a Year divisible by 4 and not 100 unless it is divisible by 400. E.g. 1800 is not a Leap Year and 2000 is a Leap Year. c. Write code having multiple if else statements based on conditions provided above and a second part having only one if statement and multiple logical

```python
year = int(input("enter the year: "))
if(year>=1582)and (year%4==0 and year%100!=0) or (year%400==0):
    print("leap year")
else:
    print("not a leap year")
```

3. Write a program to input marks and 3 subjects physics, chemistry and maths. Compute the percentage and then calculate the grade as per the following guidelines




```python
physics_marks = int(input("Enter the physics marks: "))
chemistry_marks = int(input("Enter the chemistry marks: "))
math_marks = int(input("Enter the math marks: "))

percentage = (physics_marks + chemistry_marks + math_marks) / 3

if percentage >= 80:
    print("Level 4, above agency-normalized standards")
elif 70 <= percentage <= 79:
    print("Level 3, at agency-normalized standards")
elif 60 <= percentage <= 69:
    print("Level 2, below, but approaching agency-normalized standards")
elif 50 <= percentage <= 59:
    print("Level 1, well below agency-normalized standards")
elif 40 <= percentage <= 49:
    print("Level 1, too below agency-normalized standards")
else:
    print("Remedial standards")
```

4. Write a Program to check if the given number is a prime number or not
Hint =>
a. A number that can be divided exactly only by itself and 1 are Prime Numbers,
b. Prime Numbers checks are done for numbers greater than 1
c. Loop through all the numbers from 2 to the user input number and check if the reminder is zero. If the reminder is zero break out from the loop as the number is divisible by some other number and is not a prime number.
d. Use isPrime boolean variable to store the result

```python
number = int(input("enter the number: "))
if number <=1:
    print("No, the given no is not prime")
else:
    is_prime = True
    for i in range(2, int(number**0.5)+1):
        if number % i ==0:
            is_prime = False
            break
    if is_prime:
        print("yes, the given no. is prime")
    else:
        print("No, the given no is not prime")
```

5. Write a program FizzBuzz, take a number as user input, and if it is a positive integer loop from 0 to the number and print the number, but for multiples of 3 print "Fizz" instead of the number, for multiples of 5 print "Buzz", and for multiples of both print "FizzBuzz".
Hint => a. Write the program and use for loop

```python
class FizzBuzz:
    def main():
        number = int(input("enter the number: "))
        if(number>=0):
            for i in range(0,number + 1):
                if(i%3==0 and i%5==0):
                    print("FizzBuzZ")
                if(i%3==0):
                    print("Fizz")
                elif(i%5==0):
                    print("Buzz")
FizzBuzz.main()
```

```python
class FizzBuzz:
    def main():
        number = int(input("enter the number: "))
        if(number>=0):
            i = 1
            while(i<=number):
                if(i%3==0 and i%5==0):
                    print("FizzBuzZ")
                elif(i%3==0):
                    print("Fizz")
                elif(i%5==0):
                    print("Buzz")
                else:
                    print("Buzz")
                i += 1
FizzBuzz.main()
```

7. Create a program to find the BMI of a person
Hint =>
a. Take user input in double for the weight (in kg) of the person and height (in cm) for the person and store it in the corresponding variable.
b. Use the formula BMI = weight / (height * height). Note unit is kg/m^2. For this convert cm to meter
c. Use the table to determine the weight status of the person



```python
weight = float(input("enter the weight: "))
height = float(input("enter the weight: "))
heightinMeter = height/100
BMI = weight / (heightinMeter * heightinMeter)

if(BMI>=40):
    print("Obese")
elif(BMI>=25.0 and BMI<=39.9):
    print("Overweight")
elif(BMI>=18.5 and BMI<=24.9):
    print("Normal")
else:
    print("Underweight")
```

9. Create a program to print the greatest factor of a number beside itself using a loop.
Hint =>
a. Get an integer input and assign it to the number variable. As well as define a greatestFactor variable and assign it to 1 b. Create a for loop that runs from last but one till 1 as in i = number - 1 to i = 1.
c. Inside the loop, check if the number is perfectly divisible by i then assign i to greatestFactor variable and break the loop.
d. Display the greatestFactor variable outside the loop

```python
class GreatestFactor:
    def main():
        number = int(input("enter the number: "))
        greatestFactor = 1
        for i in range(number-1, 0,-1):
            if number % i ==0:
                greatest_factor = i
                break
        print(f"The greatest factor of {number} is {greatest_factor}")
GreatestFactor.main()
```

11. Create a program to find the factors of a number taken as user input.
Hint =>
a. Get the input value for a variable named number. b. Run a for loop from i = 1 to i < number.
c. In each iteration of the loop, check if number is perfectly divisible by i.
d. If true, print the value of i.

```python
class FactorsOfNumber:
    def main():
        number = int(input("enter the number: "))
        for i in range (1, number):
            if number % i ==0:
                print(i)
FactorsOfNumber.main()
```
